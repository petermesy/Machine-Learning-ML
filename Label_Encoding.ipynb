Label Encoding
Converting labels into encoding

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

Label encofing of Breast Cancer

```python
cancer_data = pd.read_csv('/content/data.csv')
```

```python
cancer_data.head()
```

```python
# Finding amount of different labels
cancer_data['diagnosis'].value_counts()

```

```python
#Load labe; Encode function
label_encode = LabelEncoder()
```

```python
labels=label_encode.fit_transform(cancer_data.diagnosis)
```

```python
#appending the labels to datafame
cancer_data['target']=labels
```

```python
cancer_data.head()
```

```python
cancer_data['target'].value_counts()
```
