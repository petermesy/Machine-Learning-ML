Imbalanced Dataset
A dataset with an unequal class distribution

```python
#importing dependencies
import numpy as np
import pandas as pd
```

```python
credit_dataset=pd.read_csv('/content/creditcard.csv')
```

```python
credit_dataset.head()
```

```python
credit_dataset.tail()
```

```python
#Determine the distribution of two class
credit_dataset['Class'].value_counts()
```

```python
legit=credit_dataset[credit_dataset.Class==0]
fraud=credit_dataset[credit_dataset.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

Under Sapmling

Build a smple dataset containing similar didstribution and Fradulannt Transaction
NUmber of Fradulant=2

```python
legit_sample=legit.sample(n=2)
```

```python
print(legit_sample.shape)
```

Concatinate the data frames

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```
